**巢狀 if 條件敘述**
```
if condition:
  if condition:
    excute
  else:
    excute
else:
  if condition:
    excute
  elif condition:
    excute
```

````python
#  根據輸入判斷
grade = int(input("請跟我說這學期的電腦課分數："))
if grade >= 100:
  if grade > 100:
    print("超過 100 分 ？！想騙誰 ？")
  elif grade == 100:
    print("你是滿分天才！")
else:
  if grade >= 90:
    print("你非常優秀！")
  elif grade >= 60:
    print("有及格也很不錯！")
  else:
    print("再接再厲！！！")
````

**break / continue**



````python
# 印出成績清單中及格的分數
grade_list = [50, 40, 70, 80, 120, 90, 30, 100, 60, 110]
for grade in grade_list:
  if grade < 60:
    continue
  else:
    print(grade)
````

````python
# 印出成績清單中及格的分數
# 但遇到超過 100 分的成績則停止並印出錯誤訊息
grade_list = [50, 40, 70, 80, 120, 90, 30, 100, 60, 110]
for grade in grade_list:
  if grade > 100:
    print("這份成績單怪怪的喔, 有超過一百分的成績")
    break
  if grade < 60:
    continue
  else:
    print(grade)
````

````python
# 加總 1 ~ 10
# while loop
sum = 0
i = 1 
while i <= 10:
  sum = sum + i
  i = i + 1
print("加總 1 ~ 10 = %d" % (sum))
````

````python
# 加總 1 ~ 10
# while loop
sum = 0
i = 1 
while True:
  if i > 10:
    break
  sum = sum + i
  i = i + 1
print("加總 1 ~ 10 = %d" % (sum))
````

````python
def sumAtoB(a, b):
  sum = 0
  for i in range(a, b+1):
    sum += i
  return sum

sum1to10 = sumAtoB(1, 10)
sum1to100 = sumAtoB(1, 100)
print(sum1to10)
print(sum1to100)
````

**類別**

````python
class Animal():
  def __init__(self, name):
      self.name = name

a = Animal("common animal")
print(a.name)
````

````python
class Dog(Animal):
  def __init__(self, name):
    super().__init__('Dog:'+name)

a = Animal("common animal")
print(a.name)
b = Dog('Andy')
print(b.name)
````

````python
class Animal():
  def __init__(self, name):
    self.name = name
  def sound(self):
    return "not special."
 
class Dog(Animal):
  def __init__(self, name):
    super().__init__('Dog: '+name)
  def sound(self):
    return "barking"

a = Animal("common animal")
print(a.name, "的叫聲：", a.sound())
b = Dog('Andy')
print(b.name, "的叫聲：", b.sound())
````

````python
import cv2
import numpy as np

img = cv2.imread('lena.jpeg')
# 移動圖片位置 (Translation)
def translate(img, x, y):
 transMat = np.float32([[1,0,x],[0,1,y]])
 dimensions = (img.shape[1], img.shape[0])
 return cv2.warpAffine(img, transMat, dimensions)
# -x → 往左移動 ; x → 往右移動
# -y → 往上移動 ; y → 往下移動
translated = translate(img, -100, 100)
cv2.imwrite('translated_lena.jpeg', translated.astype(np.uint8))
````

````python
import cv2

img = cv2.imread('lena.jpeg')
def rotate(img, angle, rotPoint=None):
  (height,width) = img.shape[:2]
  if rotPoint is None:
    rotPoint = (width//2,height//2)
  rotMat = cv2.getRotationMatrix2D(rotPoint, angle, 1.0)
  dimensions = (width,height)
  return cv2.warpAffine(img, rotMat, dimensions)
# 輸入角度作為參數
rotated = rotate(img, -45)
cv2.imwrite('Rotated45_lena.jpeg', rotated.astype(np.uint8))
````
